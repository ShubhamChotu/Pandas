## Series with Python Functionalities

```python
# Making Series for below functions

import pandas as pd
num=[45, 7, 23, 71, 59, 31, 47, 99, 66]
alp=['A','B','C','D','E','F','G','H','I']

fun=pd.Series(num,index=alp)
print(fun)
```

### len , type , dir , sorted , min , max

```python
# len/type/dir/sorted/min/max

print('======> len function <======')
print(len(fun))   # Length of the series

print('======> type function <======')
print(type(fun))   # Type of the series

print('======> dir function <======')
print(dir(fun))   # Attributes & Methods inside the series

print('======> sorted function <======')
print(sorted(fun))   # Values in accending order

print('======> min function <======')
print(min(fun))   # Minimum value of the series

print('======> max function <======')
print(max(fun))   # Maximum value of the series
```

### type conversion

```python
# Type Conversion

Type_C=str(fun)  # list / tuple / set / dict can be done
print(Type_C)
print(type(Type_C))
```

### membership operator

```python
# Membership Operator  => search in index by-default

print(fun)
print("Search in index: ","I" in fun)

print("Search in values: ",31 in fun.values)
```

### looping

```python
# Looping  => Run all values by-default

for i in fun.head():   # values will be printed but only five because head is used
  print(i)

for j in fun.tail().index:   # index will be printed because index is used here
  print(j)
```

###  Arithmetic Operators

```python
# Arithmetic Operators  ( + , - , * , / , ** , // )

print(10 + fun.head(3))   # Add 10 no. in all values

print(100-fun.tail(3))    # no. are left out of 100

print(2*fun.head(2))      # Values are multiplied by 2

print(fun.tail(2)/2)      # Values are divided by 2

print(fun.tail(2)**2)     # power of values

print(fun.tail(2)//2)      # Values are divided by 2 in integer format
```

### Relational Operators

```python
 # Relational Operators  => ( > , >= , < , <= , !=)

print(fun.head(6) >= 45)    # greater than equal to

print(fun.tail(5) <= 20)    # less than equal to

print(fun.tail(3) != 18)    # not equal to
```

## Boolean indexing on series

```python
# Boolean indexing  => only those numbers will be printed which are above than given value

print(fun[fun >= 25])   # no. above than 25 is printed

print(fun[fun <= 30].size)  # no. of elements which are less than 30 total size
```

### Long Vs Wide Data

* Wide format is where we have a single row for every data point with multiple columns to hold the values of various attributes.

* Long format is where for each data point we have as many rows as the number of attributes and each row contains the value of a particular attribute for a given data point.

### Melt function

```python
# melt  => change wide format data to long format data
import pandas as pd

mel=pd.DataFrame(
    {
        'BCA':[120],
        'MCA':[100],
        'Phd':[75]
    }
)
mel.melt()
```

```python
# 2nd Example for melt function

ceo=pd.DataFrame({
    'Microsoft':['Satya Nadella'],
    'Google':['Sunder Pichai'],
    'Amazon':['Jeff Bezos'],
    'Twiter':['Elon Musk']
})
ceo.melt()
```

## Pivot Table
* The pivot table takes simple column-wise data as input,and groups the entires into a two-dimensional table that provides a multidimensional summarization of the data.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# This dataset is built_in in serborn library inside tips

bui=sns.load_dataset('tips')
bui.head(3)
```

```python
# find bills on the behalf of sex and smoker

# From groupby function
bui.groupby(['sex','smoker'])[['total_bill']].mean()


# From pivot_table
bui.pivot_table(index='sex',values='total_bill',columns='smoker')
```

```python
# 2nd Example for pivot table

# who pay the most tip male or female
bui.pivot_table(index='sex',columns='smoker',values='tip').unstack()
```

```python
# without values inside the pivot table

# all int columns will be printed with aggfunc
bui.pivot_table(index='sex',columns='smoker')

# To excract particular column use list
# bui.pivot_table(index='sex',columns='smoker')['tip']
```

```python
# aggfunc inside the pivot_table

# aggfunc is used to run operation on table
bui.pivot_table(index='sex',columns='smoker',values='total_bill',aggfunc='sum')
```

```python
# why separate objects to handle data and time when python already has datetime functionality?
```

* syntax wise datetime is very convenient.
* But the performance takes a hit while working with huge data.List vs Numpy array..
* The weaknesses of Python's datetime format inspired the numpy team to add a set of native time series data type to Numpy.
* The datetime64 dtype encodes as 64-bit integers,and thus allows arrays of dates to be represented very compactly.

```python
import numpy as np
date=np.array('2017-07-15',dtype=np.datetime64)
date
```

```python
date + np.arange(11)
```

* Because of the uniform type in numpy datetime64 arrays.this type of operation can be accomplished much more quickly than if we were working directly with Python's datetime objects,especially as arrays get large.
* Pandas Timestamp object combines the ease-of-use of python datetime with the efficient storage and vectorized interface of numpy.datetime64.
* From a group of these Timestamp objects,Pandas can construct a Datetimeindex that can be used to index data in a Series or DataFrame.

### to_datetime function
converts an existing objects to pandas timestamp/datatimeIndex object

```python
# simple seriess example
import pandas as pd

s=pd.Series(['2023/12/14',
             '2012/07/19',
             '2017/03/21'
             ])

# dt is the accessor
pd.to_datetime(s).dt.year  # return year
pd.to_datetime(s).dt.month  # return month
pd.to_datetime(s).dt.day  # return date
pd.to_datetime(s).dt.month_name()  # return month name
pd.to_datetime(s).dt.day_name()  # return day name
```

```python
# with error
import pandas as pd

s=pd.Series(['2023/12/14',
             '2012/122/19',
             '2017/03/21'
             ])

pd.to_datetime(s,errors='coerce').dt.month_name()
```
